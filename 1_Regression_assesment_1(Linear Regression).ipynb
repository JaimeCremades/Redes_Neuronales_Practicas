### 1.1: Plot data

```python
x_data = np.array([0.0,1.0,2.0]).reshape(3,1)
t_data = np.array([0.2,1.0,2.8]).reshape(3,1)

plt.figure(figsize=(6,4))
plt.plot(x_data,t_data,"o",markersize = 9, label = "data observations")
plt.xlabel("Altura")
plt.ylabel("Peso")
plt.title("Plot")
plt.legend()
plt.show
```

### 1.3 Compute the loss

```python
def squared_loss_function(t,y):
    
    t = np.array(t)
    y = np.array(y)
    

    sqrt_loss = np.sum((t - y) ** 2)
    
    return sqrt_loss

def absolute_loss_function(t,y):

    t = np.array(t)
    y = np.array(y)
    
    abs_loss = np.sum(np.abs(t-y))
    
    return abs_loss
```
