This is also known as logistic regression. We will simulate an example in which for each input $x$ we might have the chance with either probability $1$ or $0$ to belong to some class.

In our problem $x$ will be the weight and $y$ will represent the probability of being a dog. The other class is being a cat.

### 2.1: Plot data 

**Task**: Create numpy arrays containing the $x$ and $t$ data. This data should be represented as matrices with rows representing each data entry and columns the dimension of the input and output. Plot the data layer using matplotlib.

Use blue color (C0) and asterisc $*$ and orange color (C1) and circle $o$ to differentiate classes. 

```python
x_array = np.array([-0.13459237,-3.3015387,0.74481176,2.62434536,0.38824359,0.47182825,-0.07296862]).reshape(7,1)

t_array = np.array([0,0,0,1,1,1,1]).reshape(7,1)
```

```python
Blue = 'C0'
Orange = 'C1'

plt.figure(figsize=(6,4))
plt.plot(x_array[t_array == 0],t_array[t_array == 0],"o",color = Blue ,markersize = 9, label = "data observations class 0 cat")
plt.plot(x_array[t_array == 1],t_array[t_array == 1],"*",color = Orange ,markersize = 9, label = "data observations class 1 dog")
plt.xlabel("Peso")
plt.ylabel("Probability of being a dog")
plt.title("Plot")
plt.legend()
plt.show
```
